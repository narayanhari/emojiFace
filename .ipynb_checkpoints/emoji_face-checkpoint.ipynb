```python
import cv2
from keras.models import load_model
import numpy as np
from mtcnn.mtcnn import MTCNN
import copy
```

```python
cap = cv2.VideoCapture(0)
model = load_model("expressin_recog.h5")
detector = MTCNN()
emoji = [cv2.imread('angry.png', cv2.IMREAD_COLOR),cv2.imread('angry.png', cv2.IMREAD_COLOR),cv2.imread('fear.png', cv2.IMREAD_COLOR),cv2.imread('happy.png', cv2.IMREAD_COLOR),cv2.imread('angry.png', cv2.IMREAD_COLOR),cv2.imread('sad.png', cv2.IMREAD_COLOR),cv2.imread('surprise.png', cv2.IMREAD_COLOR)]
list = ["angry" ,"disgust","fear","happy","neutral","sad","surprise" ]
```

```python
while(1):
    status, img = cap.read()
    faces = detector.detect_faces(img)
    img2 = copy.deepcopy(img)
    x = []
    
    for face in faces:
        x1=face['box'][0]
        y1=face['box'][1]
        x2=x1+face['box'][2]
        y2=y1+face['box'][3]
        img=cv2.rectangle(img,(x1,y1),(x2,y2),[0,255,0])
        img1 = img[x1:x2 , y1:y2]
        
        dim = (face['box'][2],face['box'][3])
        
        #img2.append(img1)
    
    #for newi in img2:
        newi = cv2.resize(img1, (224,224), interpolation = cv2.INTER_AREA)
        gray = cv2.cvtColor(newi, cv2.COLOR_BGR2GRAY)
        newib = cv2.cvtColor(gray,cv2.COLOR_GRAY2BGR)
        
        new_img = np.expand_dims(newib, axis=0)
        
        k = model.predict(new_img)
        
        
        for p in k:
            m = max(p)
            for l in range(7):
                if p[l] == m:
                    e = l
                    break
        em = emoji[e]
        em = cv2. resize(em , dim, interpolation = cv2.INTER_AREA)
        
        img[y1:y2 , x1:x2] = em
        x.append(k)  
    img3  = np.append(img2 , img, axis =1)
    
    cv2.imshow("image",img3)
    if cv2.waitKey(1) == ord('e'):
        break
    '''for x1 in range(len(x[0])):
        if x[0][x1] != 0:
            print(list[x1])'''
    '''exp = []
    for a in x:
        for b in a:
            k = max(b)
            for c in range(len(b)):
                if b[c] == k:
                    exp.append(list[c])
                    break
                
    print(exp)'''
cv2.destroyAllWindows()
cap.release()
```
